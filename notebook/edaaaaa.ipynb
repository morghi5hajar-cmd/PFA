```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from datetime import datetime
```

```python
conn = sqlite3.connect("dataset/pharmacieee.db")
```

```python
query = """
SELECT
    v.id_vente,
    r.nom_region AS region,
    vi.nom_ville AS ville,
    ph.nom AS pharmacie,
    p.nom AS patient,
    m.nom AS medicament,
    m.categorie,
    v.quantite,
    v.prix_vente,
    v.date_vente,
    v.saison
FROM ventes v
JOIN patients p ON v.id_patient = p.id_patient
JOIN pharmacies ph ON v.id_pharmacie = ph.id_pharmacie
JOIN villes vi ON ph.id_ville = vi.id_ville
JOIN regions r ON ph.id_region = r.id_region
JOIN medicaments m ON v.id_medicament = m.id_medicament
"""

df = pd.read_sql_query(query, conn)
```
